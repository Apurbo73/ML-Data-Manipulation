```python
# Labeling Row and Column

df= pd.DataFrame(city,columns=['Name of cities'], index=['01','02','03'])
df
```

```python
# Merging two lists into a Zip:


city2 = ['Dhaka', 'Noakhali',' Barishal' ]
df2= pd.DataFrame(zip(city,city2),columns=["City 01", " City 02"], index=["01", "02", "03"])
df2
```

```python
df.shape
```

```python
df2.shape
```

```python
# Two Dimensional List / Nested Lists:

city3 =[["Sylhet","Majortila"], ["Chattogram","Feni"]]
df3= pd.DataFrame(city3, columns=['City','Town'], index=['01', '02'])
df3
```

```python
# Dictionary: Merging all the lists
dict1= {
    'city1': city,
    'city2': city2
}

dict1
```

```python
# Converting Dictionary into a Data frame:
dfNew= pd.DataFrame(dict1)
dfNew
```
